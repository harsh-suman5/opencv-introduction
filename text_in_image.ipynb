```python
import cv2 as cv
import os
```

```python
img_get = cv.imread(r"D:\images\aromatisse.png")
img_get = cv.resize(img_get, (500,600))

txt = cv.putText(img = img_get, 
text = "harsh academy live",
org = (50, 60),
fontFace = cv.FONT_HERSHEY_DUPLEX,
fontScale = 1,
color = (0,0,255),
thickness = 2,
lineType =cv.LINE_8,
bottomLeftOrigin = False)

txt = cv.putText(img = txt, 
text = "harsh academy live",
org = (50, 80),
fontFace = cv.FONT_HERSHEY_DUPLEX,
fontScale = 1,
color = (255,0,0),
thickness = 2,
lineType =cv.LINE_8,
bottomLeftOrigin = True)

cv.imshow("harsh academy live", txt)
cv.waitKey(0)
cv.destroyAllWindows()
```
